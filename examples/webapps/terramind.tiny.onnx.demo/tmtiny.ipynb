```python
import torch
from terratorch.registry import BACKBONE_REGISTRY

device = 'cpu'
input_size = (1, 12, 224, 224)  # batch size, channels, h, w

model = BACKBONE_REGISTRY.build('terramind_v1_tiny', pretrained=False, modalities=["S2L2A"])
model.to(device)

input = torch.rand(input_size).to(device)

# export to ONNX
torch.onnx.export(model, input, "terramind_v1_tiny.onnx", 
                  input_names=['input'], output_names=['output'], 
                  dynamic_axes={'input': {0: 'batch_size'}, 'output': {0: 'batch_size'}},
                  opset_version=14)

print('Done')
```
